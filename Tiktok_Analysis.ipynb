```python
top_hashtags=df.sort_values("posts",ascending=False).head(20)
top_hashtags
```

```python
top_per_year=df[df['year']==2025].head(20)
top_per_year
```

```python
top_peryear=df.groupby("year").apply( lambda x:x.nlargest(10,"posts"))
top_peryear
```

```python
# Display the top hashtags for years 2025 to 2022
display(top_peryear.loc[[2025, 2024, 2023, 2022]])
```

```python
import matplotlib.pyplot as plt

top_hashtags.plot(kind="bar",x="tag", y="posts")
plt.title("Top 20 Trending Hashtags on TikTok (by post count)")
plt.xlabel("Hashtags")
plt.ylabel("Posts Counts")
plt.show()
```

```python
top_per_year.plot(kind="bar", x="tag", y="posts")
plt.title("Top 20 hashtags in year 2025")
plt.xlabel("Hashtags")
plt.ylabel("Posts Counts")
plt.show()
```
